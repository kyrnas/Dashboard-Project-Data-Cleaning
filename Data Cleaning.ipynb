# Cleaning
## Covid data Country Pivot Sum

```python
import pandas as pd
```

```python
df = pd.read_csv("covid_worldwide.csv")
df
```

```python
df = pd.pivot_table(df, index=['Country_Region'],values=['Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate', 'Case_Fatality_Ratio'],aggfunc='sum')
```

```python
df.to_csv("covid_worldwide_cleaned.csv")
```

## WDIData Get desired indicators and pivot

```python
df = pd.read_csv("WDIData.csv")
df
```

```python
for i in range(1960, 2018):
    df = df.drop(columns=[str(i)])
df = df.drop(columns=[str(2021), "Unnamed: 66"])
```

```python
df = df.drop(columns=["Country Code", "Indicator Code"])
df
```

```python
query = df.loc[(df['Indicator Name'] == "Current health expenditure (% of GDP)") | (df['Indicator Name'] == "Life expectancy at birth, total (years)")]
query
```

```python
pivoted = query.pivot(index = "Country Name", columns=['Indicator Name'], values = "2018").reset_index()
pivoted
```

```python
pivoted.to_csv("WDIPivoted.csv")
```
